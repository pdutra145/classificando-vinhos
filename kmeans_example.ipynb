# KMeans Unsupervised Learning Algo

In this notebook I created fake data to train a model with k-means clustering algorithm

### Algo Summary

Basically, K means clustering is a **unsupervised learning algorithm** that try to group similar data together given a particular dataset. It constantly updates the centroids untill it finds the squared sum with the shortest distance

## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Creating fake dataset

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=250,n_features=5, centers=5, cluster_std=2.5, random_state=101)
```

```python
data[0].shape
```

```python
data[1].shape
```

```python
sns.scatterplot(data[0], )
```

## Creating and training our model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

## Visualizing and Comparing

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 6))

ax1.set_title('Original')
ax1.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap='rainbow')

ax2.set_title('Kmeans')
ax2.scatter(data[0][:, 0], data[0][:, 1], c=kmeans.labels_, cmap='turbo')
```

```python
data[1] == kmeans.labels_
```
